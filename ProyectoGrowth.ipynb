```python
import pandas as pd
import os
os.chdir('/users/yoga7i/downloads/WBPPP')
df = pd.read_csv('API_NY.GDP.PCAP.PP.CD_DS2_en_csv_v2_2546.csv', skiprows=4)
```

```python
df.head()
```

```python
# Eliminar columnas no necesarias
df.drop(columns=['Indicator Code', 'Unnamed: 69'], inplace=True)
# Eliminar columnas de años 1960 a 2019
cols_to_drop = [str(year) for year in range(1960, 2020)]
df.drop(columns=cols_to_drop, inplace=True)
df.head()
```

```python
df.to_csv('GDP_per_capita_PPP_cleaned.csv', index=False)
```

```python
# Filtrar solo los países de interés
paises = ["Mexico", "Argentina", "Brazil", "Chile", "Colombia"]
df_filtrado = df[df["Country Name"].isin(paises)]

# Transformar de formato ancho (años como columnas) a largo (Year, Value)
df_melted = df_filtrado.melt(
    id_vars=["Country Name"], 
    value_vars=["2020", "2021", "2022"],
    var_name="Year", 
    value_name="GDP per capita PPP"
)

# Crear gráfico de barras
fig = px.bar(
    df_melted,
    x="Year",
    y="GDP per capita PPP",
    color="Country Name",
    barmode="group",
    title="PIB per capita (PPA) - México vs. Argentina, Brasil, Chile y Colombia",
    labels={"GDP per capita PPP": "PIB per capita PPA (USD)", "Year": "Año"}
)
fig.update_traces(texttemplate='%{y:.7}', textposition='outside')
fig.show()
```
